```python
df_train = pd.read_csv("movie_review_train.csv")
df_test = pd.read_csv("movie_review_test.csv")
df_train.head(10)
print("Test")
df_test.head(20)
```

```python
def convert(x):
    d = {'Pos':1, 'Neg':0}
    return d[x]

df_test['label'] = df_test["class"].apply(convert)
df_test.head()
```

```python
def convert(x):
    d = {'Pos':1, 'Neg':0}
    return d[x]

df_train['label'] = df_train["class"].apply(convert)
df_train.head()
```

```python
df_test = df_test.drop('class', axis = 1)
df_test.head()
```

```python
x = df_test.text
y = df_test.label
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=1)
```

```python
#Vectorizing the sentence; removing stop words
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer (stop_words='english')
```

```python
vect.fit(x)
```

```python
vect.vocabulary_
```

```python
len(vect.vocabulary_)
dict = CountVectorizer(stop_words='english',min_df=.03, max_df=.8)
dict.fit(x)
x_train_vocabs_dict = dict.get_feature_names_out() #Latest version 
len(x_train_vocabs_dict)
```

```python
#transforming the train and test data set
x_train_transformed = vect.transform(x_train)
x_test_transformed = vect.transform(x)
```

```python
#note that the type is transformed (sparse) matrix
print(type(x_train_transformed))
print(x_train_transformed)
```

```python
from sklearn.naive_bayes import BernoulliNB
mnb = BernoulliNB()

#fit
mnb.fit(x_train_transformed,y_train)
```

```python
#predict class
y_pred_class = mnb.predict(x_test_transformed)
```

```python
#predict probabilities
y_pred_proba = mnb.predict_proba(x_test_transformed)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class[:100])
```
